Get the first 20 words from the ‚Äúnews‚Äù genre in Brown

```
import nltk
from nltk.corpus import brown

news = brown.words(categories='news')
print(news[:20])
```

**Generate bigrams from a Gutenberg text**


```
import nltk
from nltk.corpus import gutenberg
from nltk import bigrams

emma = gutenberg.words('austen-emma.txt')
emma_bigrams = list(bigrams(emma))
print(emma_bigrams[:10])
```

**Extract all words that contain at least one number**

```
import re

text = "A1 B25 S12 ahmed medhat"
words = re.findall(r'\b\w*\d+\w*\b', text)
print(words)
```

**Count the number of spaces.**

```
text = "A1 B25 S12 ahmed medhat"
count = text.count(' ')
print(count)
```

**Extract words ending in ‚Äúwd‚Äù**

```
import re

text = "A1 B25 S12 ahmed medhat sdgzgwd"
words = re.findall(r'\b\w+wd\b', text)
print(words)
```

```
import pandas as pd
from matplotlib import pyplot as plt
import sklearn.datasets

def get_iris_df():
    ds = sklearn.datasets.load_iris()
    df = pd.DataFrame(ds['data'], columns=ds['feature_names'])
    code_species_map = ds['target_names']
    df['species'] = [code_species_map[c] for c in ds['target']]
    return df

df = get_iris_df()
sums_by_species = df.groupby('species').sum()

var = 'sepal width (cm)'
sums_by_species[var].plot(kind='pie', fontsize=20)
plt.title('Breakdown for ' + var, fontsize=25)
plt.show()
```
